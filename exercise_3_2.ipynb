```python
import cv2
import matplotlib.pyplot as plt
```

```python
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
```

```python
path=r"D:\3_Color Image Processing\23.tif"
```

```python
src=cv2.imread(path)
src=cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
plt.imshow(src)
```

```python
w_1=192
h_1=128
small_size=(w_1,h_1)
```

```python
src_small=cv2.resize(src,small_size)
plt.imshow(src_small)
```

```python
w_2=768
h_2=512
big_size=(w_2,h_2)
```

```python
src_nearest=cv2.resize(src_small,big_size,interpolation=cv2.INTER_NEAREST)
```

```python
ssim_nearest = ssim(src, src_nearest,
                  win_size = 7,
                  data_range=255,
                  channel_axis=2)
psnr_nearest = psnr(src, src_nearest)
print("SSIM: ", ssim_nearest)
print("PSNR: ", psnr_nearest)
```

```python
src_bilinear=cv2.resize(src_small,big_size,interpolation=cv2.INTER_LINEAR)
```

```python
ssim_bilinear = ssim(src, src_bilinear,
                  win_size = 7,
                  data_range=255,
                  channel_axis=2)
psnr_bilinear = psnr(src, src_bilinear)
print("SSIM: ", ssim_bilinear)
print("PSNR: ", psnr_bilinear)
```

```python
src_cubic=cv2.resize(src_small,big_size,interpolation=cv2.INTER_CUBIC)
```

```python
ssim_cubic = ssim(src, src_cubic,
                  win_size = 7,
                  data_range=255,
                  channel_axis=2)
psnr_cubic = psnr(src, src_bilinear)
print("SSIM: ", ssim_cubic)
print("PSNR: ", psnr_cubic)
```

```python
plt.figure(figsize=(16,8))

plt.subplot(141)
plt.imshow(src)

plt.subplot(142)
plt.imshow(src_nearest)

plt.subplot(143)
plt.imshow(src_bilinear)

plt.subplot(144)
plt.imshow(src_cubic)
```
