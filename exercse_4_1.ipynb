```python
import cv2
import matplotlib.pyplot as plt
```

```python
path=r"D:\4_Intensity Transformation and Spatial Filtering\imx_noisy.jpg"
```

```python
src=cv2.imread(path)
```

```python
kernel_size = 5
```

```python
mean_filtered = cv2.blur(src, (5, 5))
```

```python
# 2. 应用均值滤波（内核大小5x5）
median_filtered = cv2.blur(src, (kernel_size, kernel_size))
```

```python
# 应用高斯滤波（内核大小5x5，标准差1.0；可根据需要调整sigmaX）
gaussian_filtered = cv2.GaussianBlur(src, (kernel_size, kernel_size), sigmaX=1.0)
```

```python

# 3. 可视化比较
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(src, cmap='gray')
plt.title('Original Noisy Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(mean_filtered, cmap='gray')
plt.title('Mean Filtered')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(median_filtered, cmap='gray')
#plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtered')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Filtered')
plt.axis('off')

plt.tight_layout()
plt.show()
```
