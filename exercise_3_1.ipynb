```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path=r"D:\3_Color Image Processing\7.tif"
src=cv2.imread(img_path)
#plt.subplot(131)
plt.imshow(src)
```

```python
hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

set mask red

```python
lower_red = np.array([0, 43, 46])
upper_red = np.array([10, 255, 255])
mask_red = cv2.inRange(hsv, lower_red, upper_red)
```

```python
lower_pink = np.array([156, 100, 100])
upper_pink = np.array([180, 255, 255])
mask_pink = cv2.inRange(hsv, lower_pink, upper_pink)

#  ( + )
mask = cv2.bitwise_or(mask_red, mask_pink)
```

change h value between mask value

```python
hsv_color_yellow=[]

for l in [26,28,30,32,34]:
	hsv_temp = hsv.copy()
	hsv_temp[...,0][mask > 0] = l  # Hue=30 -> yellow
	hsv_color_yellow.append(hsv_temp)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(151)
plt.imshow(cv2.cvtColor(hsv_color_yellow[0], cv2.COLOR_HSV2RGB))

plt.subplot(152)
plt.imshow(cv2.cvtColor(hsv_color_yellow[1], cv2.COLOR_HSV2RGB))

plt.subplot(153)
plt.imshow(cv2.cvtColor(hsv_color_yellow[2], cv2.COLOR_HSV2RGB))

plt.subplot(154)
plt.imshow(cv2.cvtColor(hsv_color_yellow[3], cv2.COLOR_HSV2RGB))

plt.subplot(155)
plt.imshow(cv2.cvtColor(hsv_color_yellow[4], cv2.COLOR_HSV2RGB))
plt.show()
```
